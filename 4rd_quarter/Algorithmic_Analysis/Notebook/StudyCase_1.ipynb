```python
import random
import string

class Student:
    def __init__(self, student_id, course_code, semester, grade):
        self.student_id = student_id
        self.course_code = course_code
        self.semester = semester
        self.grade = grade

# Objetivo 1: Implement sorting algorithms to organize student grade data by course, semester, or student ID for efficient storage and retrieval.
def bubble_sort_students(students, key):
    n = len(students)
    for i in range(n):
        for j in range(0, n-i-1):
            if getattr(students[j], key) > getattr(students[j+1], key):
                students[j], students[j+1] = students[j+1], students[j]

# Objetivo 2: Ordenar estudiantes por ID de su respectivo curso y mostrar el promedio de calificaciones por curso
def sort_students_by_course(students):
    # Diccionario para almacenar estudiantes agrupados por curso
    students_by_course = {}

    # Agrupar estudiantes por curso
    for student in students:
        if student.course_code not in students_by_course:
            students_by_course[student.course_code] = []
        students_by_course[student.course_code].append(student)

    # Calcular el promedio de calificaciones por curso
    for course, students_in_course in students_by_course.items():
        print(f"Students for course {course}:")
        total_grade = 0
        total_students = 0
        for student in students_in_course:
            if student.grade != 'Incomplete':
                total_grade += ord(student.grade) - ord('A') + 1  # Convertir la calificación en un valor numérico
                total_students += 1
            print(f"Student ID: {student.student_id}, Course Code: {student.course_code}, Semester: {student.semester}, Grade: {student.grade}")
        if total_students > 0:
            average_grade = total_grade / total_students
            print(f"Average grade for course {course}: {average_grade:.2f}")
        else:
            print(f"No grades available for course {course}")

# Objetivo 3: Hacer una lista e imprimir todos los estudiantes que tengan sus calificaciones incompletas
def print_students_with_incomplete_grades(students):
    incomplete_students = [student for student in students if student.grade == 'Incomplete']
    print("\nStudents with incomplete grades:")
    for student in incomplete_students:
        print(f"Student ID: {student.student_id}, Course Code: {student.course_code}, Semester: {student.semester}, Grade: {student.grade}")

# Generar datos aleatorios para estudiantes
def generate_student_data(num_students):
    students = []
    for i in range(1, num_students + 1):
        student_id = i
        course_code = random.choice(['DATOS-1', 'DATOS-2', 'CIBER-1', 'CIBER-2', 'ROBOTICS-1'])
        quarter = random.choice(['1', '2', '3'])
        grade = random.choice(['A', 'B', 'C', 'D', 'F', 'Incomplete'])
        students.append(Student(student_id, course_code, quarter, grade))
    return students

# Objetivo 4: Generar un reporte para un estudiante dado su ID
def generate_report_by_id(students, student_id):
    found = False
    for student in students:
        if student.student_id == student_id:
            print(f"Report for Student ID: {student.student_id}")
            print(f"Course Code: {student.course_code}, Semester: {student.semester}, Grade: {student.grade}")
            found = True
            break
    if not found:
        print(f"Student with ID {student_id} not found.")

# Ejecutar las funciones con los datos de los estudiantes generados
if __name__ == "__main__":
    # Generar datos de estudiantes
    students = generate_student_data(100)

    # Ordenar estudiantes por ID de su respectivo curso y mostrar el promedio de calificaciones por curso
    sort_students_by_course(students)

    # Imprimir estudiantes con calificaciones incompletas
    print_students_with_incomplete_grades(students)

    # Generar reporte para un estudiante dado su ID
    student_id = int(input("Enter student ID to generate report: "))
    generate_report_by_id(students, student_id)
```
